### Amazon Web Scraper Project 

```python
# import required modules 
import requests 
from bs4 import BeautifulSoup 
import csv 
import pandas as pd 
from itertools import zip_longest
```

```python
# amazon page url 
url = "https://www.amazon.com/s?k=laptops&crid=14BEH2D7LZI7D&sprefix=%2Caps%2C399&ref=nb_sb_ss_recent_1_0_recent"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
# use requests to fetch url 
webpage = requests.get(url , headers=headers)
# save the content of the url 
content  =webpage.content
# print(content)
```

```python
# soup object to parse the url 
soup = BeautifulSoup(content , "lxml")
# print(soup.prettify())
# get_product title and links for every product 
titles = []
products_links =[]

links = soup.find_all("a" , attrs={"class":"a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"})

for link in links: 
    titles.append(link.text.strip())
    products_links.append("https://www.amazon.com" +link.get("href"))

```

```python
# fetch all links : 
prices_list = []
rating = []
for link in products_links:
    # connect to the link 
    url_content = requests.get(link , headers= headers).content
    # create new soup to parse the content 
    soup2 = BeautifulSoup(url_content , "lxml") 

    # find the price of the product 
    price = soup2.find("span" , attrs={"class":"a-offscreen"})
    prices_list.append(price.text)

    # find the rating of the product 
    rate = soup2.find("span",attrs={"class":"a-icon-alt"})
    rating.append(rate.text)
    
```

```python
# get all data together 
file_list = [titles , prices_list , rating]
required_info = zip_longest(*file_list)
print(type(required_info))
```

```python
# create csv file to fill info 
with open("Labtops_amazon_info.csv" , "w" ,newline="") as myfile :
    wr = csv.writer(myfile)
    # header of the csv 
    wr.writerow(["Product_info" , "Price" , "Rating"])
    wr.writerows(required_info)
```

```python
# create data frame to read csv file
df = pd.read_csv("Labtops_amazon_info.csv")
```
